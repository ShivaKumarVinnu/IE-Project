## Import library, load data and select useful columns

```python
#import library

import pandas as pd
#load dataset
df = pd.read_csv('Crashes_Last_Five_Years.csv')
#filter by rural accidents
df=df[df['STAT_DIV_NAME'] == 'Country'] 
#drop unwanted columns
df.drop(['LATITUDE', 'LONGITUDE','OBJECTID','ABS_CODE', 'ACCIDENT_STATUS','VICGRID_X','VICGRID_Y', 'SRNS', 'SRNS_ALL','DIVIDED', 'DIVIDED_ALL', 'UNKNOWN', 'LGA_NAME_ALL'], axis=1, inplace=True)
print(df.columns)
print(df.shape)
```

## The dataset doesn't have LGA
#### load ACCIDENT_LOCATION.csv 
#### merge the two

```python
#merge the above with accident location to get LGA column
ACCIDENT_LOCATION = pd.read_csv('ACCIDENT/ACCIDENT_LOCATION.csv', usecols=['ACCIDENT_NO','ROAD_NAME', 'ROAD_TYPE' ])
df1 = pd.merge(df, ACCIDENT_LOCATION, on = "ACCIDENT_NO")
df1.shape
#concatenate road_name and road_type. Name it ROAD
df1['ROAD'] = df1['ROAD_NAME'] + " " + df1['ROAD_TYPE']
df1.shape
df1
```

## Group by LGA and write to csv

```python
#groupby LGA_NAME and then group by Road
groupbyLGA = df1.groupby(['LGA_NAME', 'ROAD'])
groupbyLGA = groupbyLGA['ACCIDENT_NO'].count()
#the above is now a series, convert to df
groupbyLGA=groupbyLGA.to_frame()
groupbyLGA.to_csv("aggregated_data.csv", encoding="utf-8")
groupbyLGA['ACCIDENT_NO'].sort_values
```

```python
from math import radians, sin, cos, asin, sqrt
def haversine(lon1, lat1, lon2, lat2):
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    return 2 * 6371 * asin(sqrt(a))
```

```python
CentDist = [round(haversine(row.X_x,row.Y_x,row.X_y,row.Y_y),3) for index, row in addingStuff.iterrows()]
CentDist
```
