```python
#import pandas
import pandas as pd
#Load file
AccidentLocationData = pd.read_csv('~\ACCIDENT\AccidentLocationData.csv')
#delete first column(it has serial numbers)
del AccidentLocationData['Unnamed: 0'] 
#AccidentLocationData.head()
print("Column headings:")
print(AccidentLocationData.columns)
#remove urban data
AccidentLocationDatarural = AccidentLocationData[AccidentLocationData['Deg.Urban.Name'] != 'MELB_URBAN']
AccidentLocationDatarural = AccidentLocationDatarural[AccidentLocationDatarural['Deg.Urban.Name'] != 'MELBOURNE_CBD']
#see column names
print("Column headings:")
print(AccidentLocationDatarural.columns)
#concatenate road_name and road_type. Name it ROAD
AccidentLocationDatarural['ROAD'] = AccidentLocationDatarural['ROAD_NAME'] + " " + AccidentLocationDatarural['ROAD_TYPE']
#rename Postcode.No to match the name on australian_postcodes.csv
AccidentLocationDatarural.rename(columns={'Postcode.No':'postcode'}, inplace=True)
#load australian_postcodes.csv
postcodedata = pd.read_csv('australian_postcodes.csv',usecols=['postcode','locality'])
#choose useful columns
postcodedata=postcodedata.rename(columns = {'locality':'SuburbName'})
```

```python
#merge the accident data with postcode+suburb data
AccidentLocationDatarural = pd.merge(AccidentLocationDatarural, postcodedata, on = "postcode")
```

```python
#groupby SuburbName and then group by Road
g = AccidentLocationDatarural.groupby(['SuburbName', 'ROAD'])
g = g['ACCIDENT_NO'].count()
#the above is now a series, convert to df
g=g.to_frame()
```

```python
#write to csv
g.to_csv("aggregated_data.csv", encoding="utf-8")
aggregated_data = pd.read_csv('aggregated_data.csv')
#rename last column to Accident Count
aggregated_data = aggregated_data.rename(columns={"ACCIDENT_NO": "ACCIDENT COUNT"})
aggregated_data
```
